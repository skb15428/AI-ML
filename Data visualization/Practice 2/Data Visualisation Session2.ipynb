## BoxPlot
- Numerical Data
- Quartile Distribution and Identification of Outliers

```python
# Numerical Data
# It shows quartile distribution of numerical data or numerical data vs categorical data
# Quartile = Quarter + Percentile --> 25th Percentile (Q1), 50th Percentile (Q2,median), 75th Percentile (Q3)
# Left side of box --> Q1
# Middle of box --> Q2
# Right side of box --> Q3
# Outliers will be presented by dots beyond lower and upper whiskers
# uw (Upper Whisker) = Q3+1.5*(Q3-Q1)
# lw  (Lower Whisker) = Q1-1.5*(Q3-Q1)
```

```python
df.temp.describe()
```

```python
sns.boxplot(x=df["hum"])
plt.show()
```

```python
sns.boxplot(x=df["mnth"],y=df["hum"])
plt.show()
```

## HeatMap

```python
# It shows how two or more numerical columns are co-related with each other (Correlation) (with the help of colors)
```

- Correlation -> Strength of relationship
- Range -1 to +1
- 0 to 1 -> Positive Relationship (directly proportional)
- -1 to 0 -> Negative Relationship (Inversely proportional)
- Closer to 0 but positive means weak positive relationship (0-0.3)
- Closer to 0 but negative means weak negative relationship (0-(-0.3))
- Closer to 1 means strong positive relationship (0.6-1)
- Closer to -1 means strong negative relationship (-0.6-(-1))
- If it is 0 means no relationship

```python
sns.heatmap(df[["hum","atemp","temp","cnt","windspeed"]].corr(),annot=True)
plt.show()
```

```python
# temp & cnt --> 0.627044 (Strong Positive Relationship)
# temp & atemp --> 0.991696  (Very Strong Positive Relationship)
# hum & cnt --> -0.098543 (Very weak ngative correlation or almost no correlation)
```
